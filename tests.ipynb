```python
from bs4 import BeautifulSoup
import requests
import re
```

```python
url = "https://en.wikipedia.org/wiki/Heart"
rqst = requests.get(url)
soup = BeautifulSoup(rqst.content)

paragraphs = soup.find_all("p")
paragraphs = paragraphs[:max(5, len(paragraphs))]

links = []
for e in paragraphs:
    print(e)
    links.extend(e.find_all("a"))

urls = [e['href'] for e in links if re.search("/wiki/", e['href']) and not re.search("wiktionary", e['href'])]
```
